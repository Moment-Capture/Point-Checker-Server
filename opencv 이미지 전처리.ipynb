```python
gray = cv2.imread('2021_수능 (100).jpg', cv2.IMREAD_GRAYSCALE)

img_blurred = cv2.GaussianBlur(gray, ksize=(5, 5), sigmaX=0)

img_blur_thresh = cv2.adaptiveThreshold(
    img_blurred,
    maxValue=255.0,
    adaptiveMethod=cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    thresholdType=cv2.THRESH_BINARY,
    blockSize=19,
    C=9
)

cv2_imshow(img_blur_thresh)
```

```python
import os

dirname = 'cv'
os.mkdir(dirname)

for img in os.listdir(os.getcwd()):
    gray = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
    img_blurred = cv2.GaussianBlur(gray, ksize=(5, 5), sigmaX=0)
    img_blur_thresh = cv2.adaptiveThreshold(
        img_blurred,
        maxValue=255.0,
        adaptiveMethod=cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        thresholdType=cv2.THRESH_BINARY,
        blockSize=19,
        C=9
    )
    print(img)
    cv2_imshow(img_blur_thresh)
    new_path = os.path.join(os.path.join(os.getcwd(), dirname), img)
    cv2.imwrite(new_path, img_blur_thresh)
    print()
```
