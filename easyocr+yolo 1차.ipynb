## **1차**

```python
def compute_intersect_area(rect1, rect2):

    x1, y1 = rect1[0], rect1[1]
    x2, y2 = rect1[2], rect1[3]
    x3, y3 = rect2[0], rect2[1]
    x4, y4 = rect2[2], rect2[3]

    ## case1 오른쪽으로 벗어나 있는 경우

    if x2 < x3:
        return 0

    ## case2 왼쪽으로 벗어나 있는 경우
    if x1 > x4:
        return 0

    ## case3 위쪽으로 벗어나 있는 경우
    if  y2 < y3:
        return 0

    ## case4 아래쪽으로 벗어나 있는 경우
    if  y1 > y4:
        return 0

    left_up_x = max(x1, x3)
    left_up_y = max(y1, y3)
    right_down_x = min(x2, x4)
    right_down_y = min(y2, y4)

    width = right_down_x - left_up_x
    height =  right_down_y - left_up_y

    return width * height


from ultralytics import YOLO

# Load a model
model = YOLO('./runs/detect/crop_train1_qna_2021-수능-crop_epoch50/weights/best.pt')  # pretrained YOLOv8n model

# Run batched inference on a list of images
results = model(source='../dataset/test/images/', name='crop_detect_train2_test')  # return a generator of Results objects

# Process results generator
for result in results:
    print()
    print(result.path)
    boxes = result.boxes  # Boxes object for bbox outputs
    masks = result.masks  # Masks object for segmentation masks outputs
    keypoints = result.keypoints  # Keypoints object for pose outputs
    probs = result.probs  # Probs object for classification outputs

    checked = []
    i = 0
    itr = 0

    for box in boxes:
        list_data = box.data.tolist()
        if(list_data[0][5] == 7):
            checked = list_data[0]
            itr = i
        i += 1

    j = 0
    max_size = -1
    box_size = []

    for box in boxes:
        list_data = box.data.tolist()
        if (j != itr):
            size = compute_intersect_area(checked, list_data[0])
            box_size.append(size)
            if (size > max_size):
                max_size = size
        else:
            box_size.append(0)
        j += 1

    k = 0
    for box in boxes:
        list_data = box.data.tolist()
        if (box_size[k] == max_size):
            print("고른 답")
            answer = int(list_data[0][5]-1)
            print(answer)
        k += 1
```

```python
import easyocr
import os

reader = easyocr.Reader(['ko'])
directory = '/content/drive/MyDrive/ultralytics/runs/detect/crop_detect_train1_test/crops/num'

tot_count = 0
ocr_count = 0

df.set_index('file', inplace=True)

for filename in os.scandir(directory):
    if filename.is_file():
        tot_count += 1
        result = reader.readtext(os.path.join(directory, filename), allowlist ='0123456789')
        for i in result:
            text_data = i[1]
            num = int(text_data)

            if filename.name in df.index.to_list():
                file = os.fsdecode(filename.name)
                print(file)
                df['num'].loc[[file]] = num

            if text_data != "" or text_data != " ":
                ocr_count += 1

            print(num)
        print("")

print("총 개수:", tot_count)
print("분류 개수:", ocr_count)
print("")

df.reset_index(inplace=True)

display(df)
```
